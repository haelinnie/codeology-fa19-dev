```python
# i'll  clean  all this later
scores = pd.read_csv('genome_scores.csv')
#scores.head()
#scores.isnull().any()
```

```python
tags = pd.read_csv('genome_tags.csv')
#tags.head()
#tags.isnull().any()
```

```python
movie = pd.read_csv('movie.csv')
#movie.head()
#movie.isnull().any()
```

```python
tagged = tags.merge(scores, left_on='tagId', right_on='tagId', how='inner')
tagged = movie.merge(tagged, left_on='movieId', right_on='movieId', how='inner').drop(columns=['genres'])
tagged.set_index(['movieId', 'title', 'tagId'])
```

```python
link = pd.read_csv('link.csv')
link = link.dropna()
#link.isnull().any()
link = link.astype({'tmdbId': int})
#link.head()
```

```python
tmdb = pd.read_csv('tmdb_movies_data.csv')
```
